```python
# ✅ Install dependencies
!pip install diffusers transformers accelerate safetensors torch

# ✅ Imports
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
from IPython.display import display  # <-- For showing image in Colab

# ✅ Load the model
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# ✅ Interactive Loop for Image Generation
while True:
    prompt = input("Enter your prompt (or type 'exit' to quit): ")
    if prompt.lower() == 'exit':
        print("Exiting the program.")
        break

    print("Generating image for:", prompt)
    image = pipe(prompt).images[0]

    # ✅ Display image in Colab
    display(image)

    # ✅ Save image
    image_name = prompt.replace(" ", "_")[:20] + ".png"
    image.save(image_name)
    print(f"✅ Image saved as {image_name}\n")

    # ✅ Ask user if they want another
    another = input("Do you want to generate another image? (yes/no): ")
    if another.lower() != 'yes':
        print("Exiting the program.")
        break
```
